```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("countries.csv", sep=',')
```

```python
data.head()
```

How GDP per Capita has changed over the years in Afganistan

```python
afganistan = data[data['country'] == 'Afghanistan']
```

```python
afganistan
```

```python
plt.plot(afganistan['year'], afganistan['gdpPerCapita'])
plt.show()
```

Plotting GDP of all countries against Years

```python
countries = data.country.unique()
countries_ds_dict = {}
for name in countries:
    countries_ds_dict[name] = data[data.country == name]
```

```python
for key,value in countries_ds_dict.items():
    plt.plot(value.year, value.gdpPerCapita)
    
plt.legend(countries_ds_dict.keys())
plt.show()
```
