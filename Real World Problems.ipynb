```python
kuwait = data[data.country == "Kuwait"]
kuwait
```

```python
plt.plot(kuwait.year, kuwait.gdpPerCapita)
plt.title("GDP Per Capita of Kuwait")
plt.xlabel("Years")
plt.ylabel("GDP Per Capita")
plt.show()
#GDP Per Capita dropped in 1980, what happened?
```

```python
plt.subplot(311)
plt.plot(kuwait.year, kuwait.gdpPerCapita)
plt.title("GDP Per Capita of Kuwait")
plt.xlabel("Years")
plt.ylabel("GDP Per Capita")

plt.subplot(312)
plt.plot(kuwait.year, kuwait.gdpPerCapita*kuwait.population)
plt.title("GDP of Kuwait")
plt.xlabel("Years")
plt.ylabel("GDP")

plt.subplot(313)
plt.plot(kuwait.year, kuwait.population)
plt.title("Population of Kuwait")
plt.xlabel("Years")
plt.ylabel("Population")

plt.tight_layout()
plt.show()
```

```python
#Comparing all graph
plt.plot(kuwait.year, kuwait.gdpPerCapita/kuwait.gdpPerCapita.iloc[0] * 100)
kuwait_gdp = kuwait.gdpPerCapita*kuwait.population
plt.plot(kuwait.year, kuwait_gdp/kuwait_gdp.iloc[0] * 100)
plt.plot(kuwait.year, kuwait.population/kuwait.population.iloc[0] * 100)
plt.title("GDP Per Capita, GDP, Population of Kuwait")
plt.legend(["GDP Per Capita", "GDP", "Population of Kuwait"])
plt.show()
```
