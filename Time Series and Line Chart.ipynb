```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = pd.read_csv("countries.csv", sep=",")
```

```python
countries.head()
```

```python
us = countries[countries.country == "United States"]
us.head()
```

```python
plt.title("GDP per Capita Growth in the U.S.")
plt.plot(us.year, us.gdpPerCapita)
plt.xlabel("year")
plt.ylabel("GDP per Capita")
plt.show()
```

```python
china = countries[countries.country == "China"]
china.head()
```

```python
plt.plot(us.year, us.gdpPerCapita)
plt.plot(china.year, china.gdpPerCapita)
plt.legend(['United States', 'China'])
plt.xlabel("Year")
plt.ylabel("GDP Per Capita in $")
plt.show()
```

```python
#In above graph, values are much higher and we cannot judge the Growth in GDP
#for that let's change the base value to 100.
```

```python
us_growth = us.gdpPerCapita / us.gdpPerCapita.iloc[0] * 100
china_growth = china.gdpPerCapita / china.gdpPerCapita.iloc[0]*100
```

```python
plt.plot(us.year, us_growth)
plt.plot(china.year, china_growth)
plt.legend(['United States', 'China'])
plt.title("GDP Per Capita Growth (first year = 100)")
plt.ylabel("GDP Per Capita")
plt.xlabel("Year")
plt.show()
```

```python
#Problem 2 - Population growth in US and China
plt.plot(us.year, us.population)
plt.plot(china.year, china.population)
plt.title("Population Comparison of US and China")
plt.legend(["United State", "China"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
#Relative Growth
us_population = us.population / us.population.iloc[0] / 10
china_population = china.population / china.population.iloc[0] / 10
```

```python
plt.plot(us.year, us_population)
plt.plot(china.year, china_population)
plt.title("Relative Growth in Population of US and China")
plt.legend(["United States", "China"])
plt.xlabel("Year")
plt.ylabel("Population in Billion")
plt.show()
```
