```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("countries.csv", sep=",")
```

```python
data.head(30)
```

#Fastest growing economy

```python
us = data[data.country == "United States"]
us
```

```python
countries = list(set(data.country))
```

```python
for country in countries:
    country_ds = data[data.country == country]
    y = country_ds.gdpPerCapita*country_ds.population
    y = y / y.iloc[0];
    #print(y)
    plt.plot(country_ds.year, y)
    if(countries[-1] == country):
        #plt.legend(countries)
        plt.show()
```

```python
y = data.gdpPerCapita*data.population
y = y / y.iloc[0]
data['relative_gdp'] = y
```

```python
fastest_growing = list(set(data[(data.relative_gdp > 40) & (data.year > 1980)].country))
```

```python
for country in fastest_growing:
    country_ds = data[data.country == country]
    plt.plot(country_ds.year, country_ds.relative_gdp)
    if(fastest_growing[-1] == country):
        plt.legend(fastest_growing)
        plt.show()
```

```python
data[(data.relative_gdp > 250) & (data.year < 1960)]
```

```python
data[(data.relative_gdp > 750) & (data.year > 2000)]
```

```python
us = data[data.country == "United States"]
china = data[data.country == "China"]
plt.plot(us.year, us.relative_gdp)
plt.plot(china.year, china.relative_gdp)
plt.legend(['US', 'China'])
plt.show()
```

```python
#China seems to be the Fastest growing country
```
