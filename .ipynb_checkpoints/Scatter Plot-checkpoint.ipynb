```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("countries.csv", sep=",")
```

```python
data_2007 = data[data.year == 2007]
```

```python
plt.scatter(data_2007.gdpPerCapita, data_2007.lifeExpectancy, 5)
plt.title("GDP Per Capita and Like Expectancy in 2007")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.show();
```

```python
#Checking corelations
data_2007.lifeExpectancy.corr(data_2007.gdpPerCapita)
```

```python
np.log10(data_2007.gdpPerCapita)
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita), data_2007.lifeExpectancy, 5)
plt.title("GDP Per Capita and Like Expectancy in 2007")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.show();
```

```python
#Highly Co-related
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
sorted_year = sorted(set(data.year))
```

```python
for given_year in sorted_year:
    data_year = data[data.year == given_year]
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy, 5)
    plt.xlabel("GDP per Capita")
    plt.ylabel("Life Expectancy")
    plt.title(given_year)
    plt.show()
```

```python
#Removing Outliers
for given_year in sorted_year:
    data_year = data[data.year == given_year]
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy, 5)
    plt.xlabel("GDP per Capita")
    plt.ylabel("Life Expectancy")
    plt.xlim(0, 60000)
    plt.ylim(25,85)
    plt.title(given_year)
    plt.show()
```

```python
#Finding the Outlier
data[data.gdpPerCapita > 60000]
```

```python
#Removing Outliers
for given_year in sorted_year:
    data_year = data[data.year == given_year]
    plt.scatter(np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy, 5)
    plt.xlabel("GDP per Capita")
    plt.ylabel("Life Expectancy")
    #plt.xlim(2, 5)
    plt.ylim(25,85)
    plt.title(given_year)
    plt.show()
```

```python
#Scatter plot of GDP and Life Expectancy
```

```python
data_2007["gdp"] = data_2007.population * data_2007.gdpPerCapita / 10**9
```

```python
plt.title("GDP vs Life Expectancy ")
plt.scatter(data_2007.gdp, data_2007.lifeExpectancy, 5)
plt.xlabel("GDP in Billion")
plt.ylabel("Life Expectancy")
plt.show()
```

```python
plt.title("Log GDP vs. Life Expectancy")
plt.scatter(np.log10(data_2007.gdp), data_2007.lifeExpectancy, 5)
plt.xlabel("GDP Billions log10")
plt.ylabel("Life Expectancy")
plt.show()
```

```python
#Comparing GDP vs. LifeExpectancy and GDP Per Capita vs. Life Expectancy
plt.subplot(2,1,1)
plt.title("Log GDP vs. Life Expectancy")
plt.scatter(np.log10(data_2007.gdp), data_2007.lifeExpectancy, 5)
plt.xlabel("GDP Billions log10")
plt.ylabel("Life Expectancy")


plt.subplot(2,1,2)
plt.title("Log GDP vs. Life Expectancy")
plt.scatter(np.log10(data_2007.gdpPerCapita), data_2007.lifeExpectancy, 5)
plt.xlabel("GDP Billions log10")
plt.ylabel("Life Expectancy")
plt.show()
```
